## Data Visualization 

### Distribution of Diabetes Status (0 is for no diabetes or only during pregnancy, 1 is for prediabetes, and 2 is for diabetes.)

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Diabetes_012', data=df, palette='Set2')
plt.title('Distribution of Diabetes Status')
plt.xlabel('Diabetes Status (0: No, 1: Pre-diabetes, 2: Yes)')
plt.ylabel('Count')
plt.show()
```

#### Data is imbalanced 

### Correlation Heatmap

```python
plt.figure(figsize=(14, 10))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Barplot of Average Income by Diabetes Status

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Diabetes_012', y='Income', data=df, palette='muted', ci=None)
plt.title('Average Income by Diabetes Status')
plt.xlabel('Diabetes Status (0: No, 1: Pre-diabetes, 2: Yes)')
plt.ylabel('Average Income')
plt.show()
```

### Distribution Plot of Age by Diabetes Status

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Diabetes_012'] == 0]['Age'], shade=True, label='No Diabetes', color='blue')
sns.kdeplot(df[df['Diabetes_012'] == 1]['Age'], shade=True, label='Pre-diabetes', color='orange')
sns.kdeplot(df[df['Diabetes_012'] == 2]['Age'], shade=True, label='Diabetes', color='red')
plt.title('Age Distribution by Diabetes Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
df['Age'].value_counts()
```

### Violin Plot of Mental Health by Diabetes Status

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Diabetes_012', y='MentHlth', data=df, palette='Set3')
plt.title('Mental Health Distribution by Diabetes Status')
plt.xlabel('Diabetes Status (0: No, 1: Pre-diabetes, 2: Yes)')
plt.ylabel('Number of Mentally Unhealthy Days')
plt.show()
```

### Countplot of Difficulty Walking by Diabetes Status

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='DiffWalk', hue='Diabetes_012', data=df, palette='coolwarm')
plt.title('Difficulty Walking by Diabetes Status')
plt.xlabel('Difficulty Walking (0: No, 1: Yes)')
plt.ylabel('Count')
plt.legend(title='Diabetes Status')
plt.show()
```
